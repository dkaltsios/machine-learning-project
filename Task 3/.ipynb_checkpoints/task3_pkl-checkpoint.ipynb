# Approach 1 — Threshold Extraction (White-Box)

We load the sklearn RandomForestClassifier from the `.pkl` file, extract every
split threshold for each mutable feature, and exhaustively enumerate feasible
combinations to find the **provably minimum-cost** adversarial perturbation.

```python
import pandas as pd
import numpy as np
import joblib
import pickle
from itertools import product
import warnings
warnings.filterwarnings("ignore")
```

## 1. Load data and model

```python
df = pd.read_csv("rejected_applicants.csv")
model = joblib.load("loan_approval_model.pkl")

feature_names = model.feature_names_in_ if hasattr(model, 'feature_names_in_') else None
if feature_names is None:
    with open("loan_approval_model.pkl", "rb") as f:
        feature_names = pickle.load(f)
feature_names = list(feature_names)
print("Features:", feature_names)

X = df[feature_names].copy()
```

## 2. Pick candidate applicants (highest approval probability = closest to boundary)

```python
df["approval_prob"] = model.predict_proba(X)[:, 1]
df["prediction"] = model.predict(X)

rejected = df[df["prediction"] == 0].sort_values("approval_prob", ascending=False)
print(f"Total rejected applicants: {len(rejected)}")
rejected.head(10)
```

## 3. Extract all split thresholds from the Random Forest

Each tree in the forest splits features at specific values. Between consecutive
thresholds the prediction is constant, so the optimal perturbation **must** lie
at one of these threshold boundaries.

```python
mutable_features = [
    "Requested_Loan_Amount",
    "Declared_Income",
    "Co_Applicant_Income",
    "Loan_Term_Months",
]
mutable_indices = [feature_names.index(f) for f in mutable_features]

thresholds_per_feature = {fname: set() for fname in mutable_features}

for tree in model.estimators_:
    t = tree.tree_
    for node_id in range(t.node_count):
        feat_idx = t.feature[node_id]
        if feat_idx < 0:  # leaf node
            continue
        fname = feature_names[feat_idx]
        if fname in thresholds_per_feature:
            thresholds_per_feature[fname].add(t.threshold[node_id])

for fname, vals in thresholds_per_feature.items():
    print(f"{fname}: {len(vals)} unique thresholds")
```

## 4. Cost function

```python
def calculate_cost(original, modified):
    """Return the modification cost. Assumes immutable features are unchanged."""
    cost = 0.0

    loan_decrease = original["Requested_Loan_Amount"] - modified["Requested_Loan_Amount"]
    if loan_decrease > 0:
        cost += (loan_decrease / 500) * 1

    income_increase = modified["Declared_Income"] - original["Declared_Income"]
    if income_increase > 0:
        cost += (income_increase / 1000) * 5

    co_income_increase = modified["Co_Applicant_Income"] - original["Co_Applicant_Income"]
    if co_income_increase > 0:
        cost += (co_income_increase / 1000) * 3

    term_increase = modified["Loan_Term_Months"] - original["Loan_Term_Months"]
    if term_increase > 0:
        cost += term_increase * 2

    return cost
```

## 5. Exhaustive search over threshold combinations

For each rejected applicant we:
1. Build the set of feasible candidate values per mutable feature (original value + thresholds in the allowed direction).
2. Enumerate all combinations, batch-predict, and keep the minimum-cost approved one.

```python
def find_optimal_attack(row, model, feature_names, thresholds_per_feature):
    """
    Exhaustively search all threshold combinations for a single applicant.
    Returns (best_cost, best_modified_row) or (inf, None) if no flip is possible.
    """
    orig = row[feature_names]

    # Build candidate values per mutable feature
    # Loan: can only decrease -> thresholds below original (we go just below the threshold)
    # Others: can only increase -> thresholds above original (we go just above the threshold)
    # We add a tiny epsilon to cross the threshold boundary.
    eps = 0.01

    loan_orig = orig["Requested_Loan_Amount"]
    loan_candidates = sorted(
        [loan_orig] + [th - eps for th in thresholds_per_feature["Requested_Loan_Amount"]
                       if th < loan_orig]
    )

    income_orig = orig["Declared_Income"]
    income_candidates = sorted(
        [income_orig] + [th + eps for th in thresholds_per_feature["Declared_Income"]
                         if th >= income_orig]
    )

    co_income_orig = orig["Co_Applicant_Income"]
    co_income_candidates = sorted(
        [co_income_orig] + [th + eps for th in thresholds_per_feature["Co_Applicant_Income"]
                            if th >= co_income_orig]
    )

    term_orig = orig["Loan_Term_Months"]
    term_candidates = sorted(
        [term_orig] + [th + eps for th in thresholds_per_feature["Loan_Term_Months"]
                       if th >= term_orig]
    )

    n_combos = len(loan_candidates) * len(income_candidates) * len(co_income_candidates) * len(term_candidates)
    print(f"  Candidate values — Loan: {len(loan_candidates)}, Income: {len(income_candidates)}, "
          f"CoIncome: {len(co_income_candidates)}, Term: {len(term_candidates)}")
    print(f"  Total combinations to check: {n_combos:,}")

    # Build all combinations as a matrix for batch prediction
    base_row = orig.values.copy().astype(float)
    feat_idx_loan = feature_names.index("Requested_Loan_Amount")
    feat_idx_income = feature_names.index("Declared_Income")
    feat_idx_co = feature_names.index("Co_Applicant_Income")
    feat_idx_term = feature_names.index("Loan_Term_Months")

    combos = list(product(loan_candidates, income_candidates, co_income_candidates, term_candidates))
    X_all = np.tile(base_row, (len(combos), 1))
    combo_arr = np.array(combos)
    X_all[:, feat_idx_loan] = combo_arr[:, 0]
    X_all[:, feat_idx_income] = combo_arr[:, 1]
    X_all[:, feat_idx_co] = combo_arr[:, 2]
    X_all[:, feat_idx_term] = combo_arr[:, 3]

    preds = model.predict(X_all)
    approved_mask = preds == 1

    if not approved_mask.any():
        return float("inf"), None

    # Compute cost only for approved rows
    approved_idx = np.where(approved_mask)[0]
    best_cost = float("inf")
    best_row = None

    orig_dict = {
        "Requested_Loan_Amount": loan_orig,
        "Declared_Income": income_orig,
        "Co_Applicant_Income": co_income_orig,
        "Loan_Term_Months": term_orig,
    }

    for idx in approved_idx:
        mod_dict = {
            "Requested_Loan_Amount": X_all[idx, feat_idx_loan],
            "Declared_Income": X_all[idx, feat_idx_income],
            "Co_Applicant_Income": X_all[idx, feat_idx_co],
            "Loan_Term_Months": X_all[idx, feat_idx_term],
        }
        c = calculate_cost(orig_dict, mod_dict)
        if c < best_cost:
            best_cost = c
            best_row = X_all[idx].copy()

    return best_cost, best_row
```

## 6. Run the search on all rejected applicants

```python
results = []

for rank, (idx, row) in enumerate(rejected.iterrows()):
    app_id = int(row["Applicant_ID"])
    print(f"\n=== Applicant {app_id} (rank {rank+1}, approval_prob={row['approval_prob']:.4f}) ===")
    best_cost, best_row = find_optimal_attack(row, model, feature_names, thresholds_per_feature)
    if best_row is not None:
        print(f"  ✓ Best cost: {best_cost:.2f}")
        results.append((app_id, best_cost, best_row, row))
    else:
        print(f"  ✗ No flip possible within threshold range")

results.sort(key=lambda x: x[1])
print(f"\n{'='*60}")
print(f"Global minimum cost: {results[0][1]:.2f} (Applicant {results[0][0]})")
```

## 7. Display the winning perturbation

```python
winner_id, winner_cost, winner_row, winner_orig = results[0]

orig_vals = winner_orig[feature_names].values.astype(float)
mod_vals = winner_row

comparison = pd.DataFrame({
    "Feature": feature_names,
    "Original": orig_vals,
    "Modified": mod_vals,
    "Change": mod_vals - orig_vals,
})
print(f"Applicant ID: {winner_id}")
print(f"Total cost:   {winner_cost:.2f}\n")
print(comparison.to_string(index=False))

pred = model.predict(mod_vals.reshape(1, -1))
prob = model.predict_proba(mod_vals.reshape(1, -1))[0, 1]
print(f"\nPrediction: {'APPROVED' if pred[0] == 1 else 'REJECTED'}")
print(f"Approval probability: {prob:.4f}")
```
